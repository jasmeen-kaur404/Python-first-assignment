```python
# Compare and contrast mutable and immutable objects in Python with examples?

In Python, objects can be either changeable (modifiable) or unchangeable (unchangeable). 
Mutable objects, such as lists and dictionaries, can be modified after creation, while immutable objects, such as tuples and strings, cannot. 
Mutability affects how objects are stored and manipulated in memory, impacting performance, memory usage, and concurrency in Python programs. 


Mutable objects
Examples:
Lists: You can add, remove, or change items.
my_list = [1, 2, 3]
my_list.append(4)  # Now the list is [1, 2, 3, 4]

Dictionaries: You can update values or add new key-value pairs.
my_list = [1, 2, 3]
my_list.append(4)  # Now the list is [1, 2, 3, 4]


Immutable objects
Examples:
Strings: Creating a new string is necessary to make changes
my_string = "hello"
new_string = my_string.upper()  # new_string is "HELLO", original is unchanged

Tuples: Elements cannot be modified after creation.
my_tuple = (1, 2, 3)
# You need to create a new tuple to change values, like (4, 5, 6)
```

```python
# Discuss the different types of operators in Python and provide examples of how they are used?

Operators are special symbols or keywords that are used to carry out specific actions on numbers or variables in Python expressions. 
Python supports various types of operators, including arithmetic operators (+, -, *, /), comparison operators (==, !=, <, >), logical operators (and, or, not), assignment operators (=, +=, -=, *=, /=), etc.
Operators have precedence and associativity rules that determine the order of evaluation in expressions.


1. Arithmetic Operators
These operators perform basic mathematical operations.

Addition (+): Adds two values.
a = 5
b = 3
result = a + b  # result is 8

Subtraction (-): Subtracts one value from another.
result = a - b  # result is 2


Multiplication (*): Multiplies two values.
result = a * b  # result is 15


Division (/): Divides one value by another (returns a float).
result = a / b  # result is 1.666...


Floor Division (//): Divides and returns the largest integer less than or equal to the result.
result = a // b  # result is 1

Modulus (%): Returns the remainder of a division.
result = a % b  # result is 2


Exponentiation (**): Raises one value to the power of another.
result = a ** b  # result is 125



2. Comparison Operators
These operators compare values and return True or False.

Equal to (==): Checks if two values are equal.
result = (a == b)  # result is False


Not equal to (!=): Checks if two values are not equal.
result = (a != b)  # result is True


Greater than (>): Checks if one value is greater than another.
result = (a > b)  # result is True


Less than (<): Checks if one value is less than another.
result = (a < b)  # result is False


Greater than or equal to (>=): Checks if one value is greater than or equal to another.
result = (a >= b)  # result is True


Less than or equal to (<=): Checks if one value is less than or equal to another.
result = (a <= b)  # result is False



3. Logical Operators
These operators are used to combine conditional statements.

And (and): Returns True if both conditions are true.
result = (a > 0 and b > 0)  # result is True


Or (or): Returns True if at least one condition is true.
result = (a > 0 or b < 0)  # result is True


Not (not): Returns True if the condition is false.
result = not(a > b)  # result is True



4. Assignment Operators
These operators are used to assign values to variables.

Assign (=): Assigns a value to a variable.
x = 10


Add and assign (+=): Adds a value to a variable and assigns the result.
x += 5  # Equivalent to x = x + 5, x is now 15


Subtract and assign (-=): Subtracts a value from a variable and assigns the result.
x -= 3  # Equivalent to x = x - 3, x is now 12


Multiply and assign (*=): Multiplies a variable by a value and assigns the result.
x *= 2  # Equivalent to x = x * 2, x is now 24


Divide and assign (/=): Divides a variable by a value and assigns the result.
x /= 4  # Equivalent to x = x / 4, x is now 6.0


Modulus and assign (%=): Applies modulus and assigns the result.
x %= 4  # Equivalent to x = x % 4, x is now 2.0


Exponentiation and assign (**=): Raises a variable to a power and assigns the result.
x **= 2  # Equivalent to x = x ** 2, x is now 4
```

```python
# Explain the concept of type casting in Python with examples?

Type casting, also referred to as type conversion, is the process of changing one data type to another in Python. 
Python provides built-in functions for type casting, such as int(), float(), str(), list(), tuple(), dict(), etc.
Type casting is often necessary for performing arithmetic operations, data manipulation, and input/output operations in Python programs.


1. Casting to Integer (int)
Description: Converts a value to an integer. This can be done from a float or a string that represents a whole number.

From Float to Integer:
num_float = 12.75
num_int = int(num_float)  # Converts float to integer, drops the decimal part
print(num_int)            # Output: 12

From String to Integer:
num_str = "42"
num_int = int(num_str)  # Converts string to integer
print(num_int)          # Output: 42


2. Casting to Float (float)
Description: Converts a value to a float. This can be done from an integer or a string that represents a number with or without decimal points.

From Integer to Float:
num_int = 7
num_float = float(num_int)  # Converts integer to float
print(num_float)            # Output: 7.0


From String to Float:
num_str = "3.14"
num_float = float(num_str)  # Converts string to float
print(num_float)            # Output: 3.14



3. Casting to String (str)
Description: Converts a value to a string. This can be done from integers, floats, or other types.

From Integer to String:
num_int = 123
num_str = str(num_int)  # Converts integer to string
print(num_str)         # Output: "123"


From Float to String:
num_float = 45.67
num_str = str(num_float)  # Converts float to string
print(num_str)           # Output: "45.67"



4. Casting to List (list)
Description: Converts a sequence or iterable to a list.

From Tuple to List:
num_tuple = (1, 2, 3)
num_list = list(num_tuple)  # Converts tuple to list
print(num_list)             # Output: [1, 2, 3]


From String to List:
num_str = "hello"
num_list = list(num_str)  # Converts string to list of characters
print(num_list)          # Output: ['h', 'e', 'l', 'l', 'o']


5. Casting to Tuple (tuple)
Description: Converts a sequence or iterable to a tuple.

From List to Tuple:
num_list = [1, 2, 3]
num_tuple = tuple(num_list)  # Converts list to tuple
print(num_tuple)             # Output: (1, 2, 3)


From String to Tuple:
num_str = "abc"
num_tuple = tuple(num_str)  # Converts string to tuple of characters
print(num_tuple)           # Output: ('a', 'b', 'c')


6. Casting to Set (set)
Description: Converts a sequence or iterable to a set, which is an unordered collection of unique elements.

From List to Set:
num_list = [1, 2, 2, 3]
num_set = set(num_list)  # Converts list to set, duplicates are removed
print(num_set)          # Output: {1, 2, 3}


From String to Set:
num_str = "hello"
num_set = set(num_str)  # Converts string to set of unique characters
print(num_set)         # Output: {'h', 'e', 'l', 'o'}

```

```python
# Describe the different types of loops in Python and their use cases with examples.

In Python, loops are employed to repeat a sequence of actions or code until a specific condition is fulfilled. 
Python offers two primary types of loops: for loops and while loops. 
“for” loops are used for iterating over a sequence of elements, while “while” loops are used for executing code until a specified condition becomes False

1. for Loop
The for loop is used to iterate over a sequence (like a list, tuple, string, or range). It is ideal for iterating over elements of a collection or a sequence.

Use Cases:
Iterating through elements in a list or tuple.
Iterating through characters in a string.
Repeating actions a specific number of times using range().

Example:

Iterating Over a List:
    fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
    
#output: 
apple
banana
cherry

Using range() to Repeat Actions:
    for i in range(5):
    print(i)
#output: 
0
1
2
3
4


2. while Loop
Description: The while loop repeatedly executes a block of code as long as its condition is True. It is useful when the number of iterations is not known beforehand and depends on dynamic conditions.

Use Cases:

Continuously reading user input until a valid response is received.
Performing operations until a specific condition is met.
Example:

Basic While Loop:

python
Copy code
count = 0

while count < 5:
    print(count)
    count += 1
Output:

Copy code
0
1
2
3
4

```
